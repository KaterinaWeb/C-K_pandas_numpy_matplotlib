```python
merged_df = pd.merge(df_customers, df_orders, on = 'customer_id')
```

```python
df_not_delivered = merged_df.query('order_status != "delivered"') \
                .reindex(['customer_unique_id','order_id', 'order_status', 'creating_order'], axis = 1)
df_not_delivered['order_month'] = df_not_delivered['creating_order'].dt.to_period('M')
df_not_delivered.head()
```

```python
df_count_not_delivered_orders = df_not_delivered.groupby(['order_month', 'order_status'], as_index = False) \
                .agg({'order_id' : 'nunique'}) \
                .rename(columns = {'order_id' :'count_not_delivered_orders'})
                
df_count_not_delivered_orders.head()               
```

```python
#посчитаем среднее количество недоставленных заказов с группировкой по статусу
df_result = df_count_not_delivered_orders \
                .groupby('order_status', as_index = False) \
                .agg({'count_not_delivered_orders' : 'mean'}) \
                .rename(columns = {'count_not_delivered_orders': 'mean_not_delivered_orders'}) \
                .sort_values('mean_not_delivered_orders', ascending = False) 
df_result
```

```python
#Создаем таблицу возможных причин
df_reasons = pd.DataFrame({'order_status': ['shipped', 'unavailable', 'canceled', 'processing', 'invoiced',
       'created', 'approved'], 'reason_for_not_delivering': ['потеряно при отгрузке',
                                                             'заказ больше недоступен', 
                                                             'заказ отменен пользователем', 
                                                             'потеряно на складе в процессе сборки заказа',
                                                             'потеряно на стадии invoiced', 
                                                             'заказ создан, но не оплачен',
                                                             'потеряно на стадии approved']})
df_reasons
```

```python
reason_table = pd.merge(df_result, df_reasons, on = 'order_status')
reason_table
```

```python
df_answer = reason_table.reindex(['reason_for_not_delivering', 'mean_not_delivered_orders'], axis = 1)
```

```python
#5. Используя pandas, проведи когортный анализ пользователей. 
#В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.
```

```python
#Давайте проанализируем когортный retention на 3-й месяц только для тех пользователей, которые совершили покупку.
#Под покупкой, как и прежде, я буду понимать товары, которые были оплачены и не отменены, а также товары со статусом delivered.
```

```python
merged_df = pd.merge(df_customers, df_orders, on = 'customer_id')
merged_df = pd.merge(merged_df, df_order_items, on = 'order_id')
merged_df = merged_df.reindex(['customer_unique_id', 'order_status', 'creating_order', 'order_item_id', 'price'], axis = 1)
mask = (merged_df.order_status == 'delivered') | ((merged_df.order_status != 'canceled') & (merged_df.creating_order.notnull()))
merged_df = merged_df[mask]
```

```python
merged_df['order_month'] = merged_df['creating_order'].dt.to_period('M')
merged_df.head()
```

```python
#Создадим столбец с месяцем первого заказа каждого пользователя (когортой)
cohort_users = merged_df.groupby(['customer_unique_id', 'order_month'], as_index = False).agg({'order_item_id':'count'}) \
    .sort_values(['customer_unique_id', 'order_month']) \
    .groupby('customer_unique_id') \
    .head(1) \
    .rename(columns={'order_month':'first_purchase'})
cohort_users = cohort_users.reindex(['customer_unique_id', 'first_purchase'], axis = 1)
cohort_users.head()
```

```python
#Найдем дату (год-месяц) следующей покупки для рассчета возврата клиентов через 3 месяца после 1 заказа.
#Прибавим 3 месяца к дате первой покупки
cohort_users['first_purchase'] = cohort_users['first_purchase'].dt.to_timestamp()
cohort_users['first_purchase'] = pd.to_datetime(cohort_users['first_purchase'])
cohort_users['next_purchase'] = cohort_users['first_purchase'] + pd.DateOffset(months=3)
```

```python
#Вернем к формату год-месяц
cohort_users['next_purchase'] = cohort_users['next_purchase'].dt.to_period('M')
cohort_users['first_purchase'] = cohort_users['first_purchase'].dt.to_period('M')
cohort_users.head()
```

```python
#Находим лояльных клиентов, которые совершили покупки, через 3 месяца после 1 покупки
loyal_clients = pd.merge(merged_df, cohort_users, left_on=['customer_unique_id', 'order_month'], 
                         right_on = ['customer_unique_id', 'next_purchase'])
loyal_clients = loyal_clients.reindex(['customer_unique_id', 'first_purchase', 'next_purchase'], axis = 1)
loyal_clients.head()
```

```python
#Найдем количество лояльных клиентов по когортам
loyal_clients = loyal_clients.groupby(['next_purchase', 'first_purchase'], as_index = False).customer_unique_id.nunique() \
    .rename(columns={'customer_unique_id': 'count_loyal_customers'})
loyal_clients.head()
```

```python
#Рассчитаем размер каждой когорты на старте, т.е. количество уникальных пользователей в когорте на старте
cohort_sizes = merged_df.groupby(['order_month'], as_index=False)['customer_unique_id'].nunique() \
    .rename(columns={'customer_unique_id': 'count_total_customers'})
cohort_sizes.head()
```

```python
# Смерджим последние таблицы и посчитаем retention rate
df_retention = pd.merge(loyal_clients, cohort_sizes, left_on = 'first_purchase', right_on = 'order_month')
df_retention = df_retention.reindex(['order_month', 'count_total_customers', 'next_purchase', 'count_loyal_customers'], axis = 1)
df_retention = df_retention.assign(CRR = df_retention.count_loyal_customers / df_retention.count_total_customers * 100)
df_retention.sort_values('CRR', ascending = False)
```

```python
#Визуализируем CRR через boxplot, посмотрим медину, межквартильный размах, наличие выбросов
plt.figure(figsize=(8, 5))
sns.boxplot(df_retention.CRR, color = 'darkgreen')
sns.despine()

plt.show()
```

```python
df_retention.dtypes
```

```python
table = df_retention.reindex(['order_month', 'CRR'], axis = 1)
table['order_month'] = table['order_month'].dt.to_timestamp()
table['order_month'] = pd.to_datetime(table['order_month'])
table.head()
```

```python
plt.figure(figsize = (8,5))
sns.lineplot(data = table, x = 'order_month', y = 'CRR')
sns.despine()
plt.xlabel('\n Дата заказа', fontsize = 15) 
plt.ylabel('CRR', fontsize = 15)
plt.title('Retention покупателей на 3 месяц \n', fontsize = 20, color = 'brown') 
```

```python
#6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, 
#построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать 
#следующие метрики: R - время от последней покупки пользователя до текущей даты, 
#F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. 
#Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик 
#recency, frequency и monetary для интерпретации этих кластеров.
```

```python
#Меня интересует, сколько раз клиент покупал, а не сколько товаров он купил.
#Будем считать заказы, а не товары.
```

```python
#Будем делать RFM-сегментацию пользователей. Под пользователем в данном случае разумно считать только тех, кто совершил покупку.
#Покупка, как и прежде, товары, которые были оплачены и не отменены, а также товары со статусом delivered. 
```

```python
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python
merged_df = pd.merge(df_customers, df_orders, on = 'customer_id')
merged_df = pd.merge(merged_df, df_order_items, on = 'order_id')
mask = (merged_df.order_status == 'delivered') | ((merged_df.order_status != 'canceled') & (merged_df.creating_order.notnull()))
merged_df = merged_df[mask]
merged_df = merged_df.reindex(['customer_unique_id', 'order_id', 'order_status', 'creating_order', 'price'], axis = 1)
merged_df.head()
```

```python
#В примере из задания про RFM-сегментацию сказано : We have about one year of sales data (from December 2010 to December 2011).
#This is the period usually used for RFM analysis. I'll set the period of the study to 365 days to get exactly one year.- , 
#что чаще всего используется календарный год для анализа. Следовательно, в задании мы тоже можем выбрать определенный год 
#и делать RFM-сегментацию по нему.
```

```python
merged_df = merged_df.assign(year = merged_df.creating_order.dt.year)
new = merged_df[merged_df.year == 2017]
```

```python
# Возьмем 2017 год, так как по нему есть  данные за целый год с января по декабрь.
print('Orders from {} to {}'.format(new.creating_order.min(),
                                    new.creating_order.max()))
```

```python
#Посчитаем сумму каждого заказа
orders = new.groupby(['order_id', 'creating_order', 'customer_unique_id'], as_index = False).agg({'price': 'sum'})
orders.head()
```

```python
#Cмоделируем анализ,установив дату СЕЙЧАС на один день после последней покупки.
#Эта дата будет использоваться в качестве ориентира для расчета оценки недавности.
```

```python
NOW = orders['creating_order'].max() + timedelta(days=1)
NOW
```

```python
orders['DaysSinceOrder'] = orders['creating_order'].apply(lambda x: (NOW - x).days)
orders.head()
```

```python
#Построим таблицу RFM для каждого клиента.
```

```python
aggr = {
    'DaysSinceOrder': lambda x: x.min(),  # количество дней с момента последнего заказа (Recency)
    'creating_order': lambda x: len([d for d in x]), #общее количество заказов за последний период (Frequency)
}
rfm = orders.groupby('customer_unique_id', as_index = False).agg(aggr)
rfm.rename(columns={'DaysSinceOrder': 'Recency', 'creating_order': 'Frequency'}, inplace=True)
rfm.head()
```

```python
# Добавим Monetary - сколько денег мы заработали с каждого клиента за 2017 год.
rfm['Monetary'] = rfm['customer_unique_id'].apply(lambda x: orders[(orders['customer_unique_id'] == x)]\
                                                           ['price'].sum())
rfm.head()
```

```python
rfm.describe()
```

```python
quintiles = rfm[['Recency', 'Frequency', 'Monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
rec_bins = [0, 40, 90, 150, 250, 365]
rec = pd.cut(rfm.Recency, rec_bins, labels = ['5', '4', '3', '2', '1'])
rfm['R'] = pd.Series(rec)

freq_bins = [0, 1, 3, 5, 6, 9]
freq = pd.cut(rfm.Frequency, freq_bins, labels = ['1', '2', '3', '4', '5'])
rfm['F'] = pd.Series(freq)

mon_bins = [0, 40, 70, 110, 200, 15000]
mon = pd.cut(rfm.Monetary, mon_bins, labels = ['1', '2', '3', '4', '5'])
rfm['M'] = pd.Series(mon)
```

```python
Classter_table = pd.DataFrame({'Category_Recency': ['251-365', '151-250', '91-150', '41-90', '0-40'], 
              'Category_Frequency': ['0-1', '2-3', '4', '5', '6-9'],
              'Category_Monetary': ['0-40', '41-70', '71-110', '111-200', '201-15000'],
              'Classter': ['1', '2', '3', '4', '5']})
Classter_table
```

```python
rfm['RFM Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
```

```python
#Визуализируем сегменты наших клиентов
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))

for i, p in enumerate(['R', 'F', 'M']):
    parameters = {'R':'Recency', 'F': 'Frequency', 'M':'Monetary'}
    y = rfm[p].value_counts().sort_index()
    x = y.index
    ax = axes[i]
    bars = ax.bar(x, y, color='silver')
    ax.set_frame_on(False)
    ax.tick_params(left=False, labelleft=False, bottom=False)
    ax.set_title('Distribution of {}'.format(parameters[p]),
                fontsize=14)
    for bar in bars:
        value = bar.get_height()
        if value == y.max():
            bar.set_color('firebrick')
        ax.text(bar.get_x() + bar.get_width() / 2,
                value - 5,
                '{}\n({}%)'.format(int(value), int(value * 100 / y.sum())),
               ha='center',
               va='top',
               color='w')

plt.show()
```

```python
# визуализируем процент клиентов в каждом сегменте
segments_counts = rfm['Segment_Description'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(10, 8))

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['Low_champions', 'Gold_champions', 'Middle_champions', 
                                        'Gold_potential loyalists', 'Middle_potential loyalists']:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )

plt.show()
```
